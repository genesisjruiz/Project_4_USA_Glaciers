```python
import pandas as pd
```

```python
dashboard_df_1 = pd.read_csv('../resources/us_glaciers_data_1.csv')

dashboard_df_1
```

```python
dashboard_df_1 = dashboard_df_1.drop(columns=['Unnamed: 0', 'area', 'glac_id', 'min_elev', 'mean_elev', 'max_elev', 'src_date', 'release_dt', 'first_coordinate', 'cluster', 'closest_cluster_lat', 'closest_cluster_lng'])

dashboard_df_1
```

```python
line_type = {
    'dc': 'Debris Covered',
    'gb': 'Glacier Bound',
    'ir': 'Internal Rock'
}

dashboard_df_1['line_type'] = dashboard_df_1['line_type'].replace(line_type)

dashboard_df_1
```

```python
glac_stat = {
    'e': 'Exists',
    'g': 'Gone'
}

dashboard_df_1['glac_stat'] = dashboard_df_1['glac_stat'].replace(glac_stat)

dashboard_df_1
```

```python
gone_date = {
    'e': 'N/A'
}

dashboard_df_1['gone_date'] = dashboard_df_1['gone_date'].replace(gone_date)

dashboard_df_1
```

```python
dashboard_df_1 = dashboard_df_1[['line_type', 'anlys_time', 'db_area', 'glac_name', 'glac_stat', 'gone_date', 'lat', 'lng']]

dashboard_df_1
```

```python
dashboard_df_1['anlys_time'] = pd.to_datetime(dashboard_df_1['anlys_time'], unit='s')

dashboard_df_1
```

```python
dashboard_df_1 = dashboard_df_1[~((dashboard_df_1['glac_name'].str.len() == 2) & (dashboard_df_1['glac_name'].str.isalpha()) | 
          dashboard_df_1['glac_name'].str.isdigit())]

dashboard_df_1
```

```python
dashboard_df_2 = dashboard_df_1.copy()

dashboard_df_2

dashboard_df_2['Area Analysis Date'] = dashboard_df_2['anlys_time'].dt.date
```

```python
dashboard_df_2 = dashboard_df_2.drop(columns=['anlys_time'])

dashboard_df_2 = dashboard_df_2.rename(columns={'line_type': 'Glacier Type', 'Analysis Time': 'Analysis Time', 'db_area': 'Area',
                                               'glac_name': 'Glacier Name', 'glac_stat': 'Glacier Exists?', 'gone_date': 'Glacier Melt Record Date', 
                                                'lat': 'Latitude', 'lng': 'Longitude'}) 

dashboard_df_2
```

```python
dashboard_df_2 = dashboard_df_2[['Glacier Name', 'Glacier Exists?', 'Area', 'Glacier Melt Record Date',
                                'Latitude', 'Longitude', 'Area Analysis Date']]

dashboard_df_2
```

```python
dashboard_df_3 = dashboard_df_2.copy()

# Group by 'glacier name' and get the first value for 'latitude' and 'longitude'
first_values = dashboard_df_3.groupby('Glacier Name').first().reset_index()

# Merge the first values back to the original DataFrame, keeping only 'glacier name'
dashboard_df_3 = pd.merge(dashboard_df_3[['Glacier Name']], first_values, on='Glacier Name', how='left')

dashboard_df_3['Area Analysis Date'] = pd.to_datetime(dashboard_df_3['Area Analysis Date'])

# Extract year and month and set day to the first of the month and time to midnight
dashboard_df_3['Area Analysis Date'] = dashboard_df_3['Area Analysis Date'].apply(
    lambda x: pd.Timestamp(year=x.year, month=x.month, day=1)
)

dashboard_df_3
```

```python
temperature_data = pd.read_csv('../resources/Monthly_Avg_Temp_1990_2024.csv')

temperature_data
```

```python
temperature_data['formatted_date'] = temperature_data['Date'].apply(lambda x: f"{str(x)[:4]}-{str(x)[4:]}")

temperature_data['datetime'] = pd.to_datetime(temperature_data['formatted_date'], format='%Y-%m')

temperature_data
```

```python
temperature_data = temperature_data.drop(columns=['Date', 'Anomaly', 'formatted_date'])

temperature_data
```

```python
# Merge the DataFrames on the datetime columns
dashboard_df_3 = pd.merge(dashboard_df_3, temperature_data, left_on='Area Analysis Date', right_on='datetime', how='left')

# Display the resulting DataFrame
dashboard_df_3
```

```python
dashboard_df_3 = dashboard_df_3.drop(columns=['datetime'])

dashboard_df_3 = dashboard_df_3.rename(columns={'Value': 'Temperature in F'})

dashboard_df_3
```

```python
dashboard_df_3.to_csv('../resources/dashboard_df.csv')
```
